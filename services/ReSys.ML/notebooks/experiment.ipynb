# ML Service Experimentation
This notebook demonstrates how to load the model locally and test embeddings.

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = ['Classic White T-Shirt', 'Modern Red Sneaker', 'Vintage Denim Jacket']
embeddings = model.encode(sentences)

print(f"Embedding shape: {embeddings.shape}")
print(f"Cosine similarity (0 vs 1): {np.dot(embeddings[0], embeddings[1])}")
```
